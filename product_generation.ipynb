```python
import torch
from PIL import Image
from transformers import BlipProcessor, BlipForConditionalGeneration

# THIS CODE BLOCK IS FOR DRAWING TO PRODUCT GENERATION

model_name = "Salesforce/blip-image-captioning-large" # Use 'large' for potentially better detail
processor = BlipProcessor.from_pretrained(model_name)
model = BlipForConditionalGeneration.from_pretrained(model_name)

device = "cuda" if torch.cuda.is_available() else "cpu"
model.to(device)

image_path = "/content/dress_drawing.png"
image = Image.open(image_path).convert("RGB")

inputs = processor(images=image, return_tensors="pt").to(device)

generated_ids = model.generate(
    pixel_values=inputs.pixel_values,
    max_new_tokens=75,          # Allow for longer captions
    num_beams=5,                # Use beam search for better quality
    length_penalty=1.5,         # Encourage slightly longer output
    no_repeat_ngram_size=3,     # Avoid repeating 3-word sequences
    early_stopping=True
)

generated_caption = processor.decode(generated_ids[0], skip_special_tokens=True)
print(f"Generated Caption: {generated_caption}")

import torch
from diffusers import DiffusionPipeline
from accelerate import Accelerator # Import the Accelerator class

# 1. Instantiate the Accelerator
accelerator = Accelerator()

# Load Previous Pipeline
pipeline = DiffusionPipeline.from_pretrained(
    "stabilityai/stable-diffusion-2", revision=None, variant=None, torch_dtype=torch.float32
)

# 2. Use the device from the *instance* of Accelerator
pipeline = pipeline.to(accelerator.device)

# Load attention processors
pipeline.unet.load_attn_procs("NouRed/sd-fashion-products")

# Run Inference
# 3. Use the device from the *instance* of Accelerator for the generator as well
generator = torch.Generator(device=accelerator.device)
seed = 42
if seed is not None:
    generator = generator.manual_seed(seed)

# Ensure generated_caption is defined or passed correctly
# For this example, let's assume it's defined.
# If it comes from another script/module, make sure it's imported correctly.
# For a standalone test, you might define it here:

prompt = "Generate only the picture of the product kept on a plain white background: " + generated_caption
image = pipeline(prompt, num_inference_steps=40, generator=generator).images[0]

# Save Generated Product
image.save("generated_img.png")

print(f"Image saved to generated_img.png")

```
